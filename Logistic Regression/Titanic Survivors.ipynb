```python
## Now to get the dummies 
sex = pd.get_dummies(titanic['Sex'], drop_first=True )
sex
```

```python
embark = pd.get_dummies(titanic['Embarked'],drop_first=True)
embark
```

```python
pclass = pd.get_dummies(titanic['Pclass'] , drop_first=True)
pclass
```

```python
titanic = pd.concat([titanic,sex,embark,pclass], axis=1)
```

```python
titanic.head()
```

```python
titanic.drop(['Name','Sex','Name_wiki','Boarded','Destination','Hometown'],axis=1 , inplace=True)
```

```python
titanic.head(5)
```

```python
titanic.drop(['Ticket','Embarked'] , axis=1 , inplace=True)
```

```python
titanic.head()
```

```python
titanic.drop(['Pclass','PassengerId'] , axis=1 , inplace=True)
```

```python
titanic.info()
```

```python
### now droppinng the survived column
X = titanic.drop('Survived',axis=1)
y = titanic['Survived']
```

```python

titanic.head(5)
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size=0.3 , random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression(multi_class="auto",max_iter=15000)
```

```python
logmodel.fit(X_train , y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```
